```python
from bs4 import BeautifulSoup
import csv
import re
import pandas as pd 
```

```python
html = open("realestate.htm").read()
soup = BeautifulSoup(html)
```

```python
addr_code_soup = soup.find_all('address', attrs={'class': 'list-card-addr'})
```

```python
len(addr_code_soup)
# Address
address = []
for addr in addr_code_soup:
    address.append(addr.getText())
```

```python
heading_code_soup = soup.find_all('div', attrs={'class': 'list-card-heading'})
```

```python
price = []
attribute = []
count = 0
row = {}
```

```python
for head in heading_code_soup:
    for body in head:
        if count % 2 == 0:
            price.append(body.getText())
        else:
            attribute.append(body.getText())
        count = count + 1
```

```python
bed = []
bath = []
size = []
for attr in attribute:
    attrArray = [int(s) for s in re.findall(r'-?\d+\.?\d*', attr)]
    bed.append(attrArray[0]) # bed
    bath.append(attrArray[1]) # baths
    if len(attrArray) > 3:
#         print("adding size 1000")
        size.append( attrArray[2] * 1000 + attrArray[3]) #sqft
    else:
#         print("adding size sqft")
        if attrArray[2] < 11:
            size.append(attrArray[2] * 43560) # seqft
        else:
            size.append(attrArray[2]) #sqft
    
```

```python
len(size)
```

```python
row = {"addr":address,"price":price,"bed":bed,"bath":bath,"size":size}
```

```python
df = pd.DataFrame(row)
```
